```python
import pandas as pd
import json
```

```python
medrt_cui = {}

for line in open('/Users/dgg32/Documents/drug_atc/neo4j_files/moa_12.json', 'r'):
  obj = json.loads(line)

  for m in obj["medrt_id"]:
    medrt_cui[m] = obj["id"]
```

```python
medrt_cui["N0000000153"] = "C1373072"
medrt_cui["N0000194042"] = "C5781807"
medrt_cui["N0000184150"] = "C3267923"
medrt_cui["N0000178377"] = "C2917239"
```

```python
moa_df = pd.read_csv("drugs/moa_8.tsv", sep="\t")
```

```python
moa_df.head()
```

```python
moa_df['id'] = moa_df['id'].map(medrt_cui)
```

```python
unmapped = moa_df[moa_df['id'].isna()]
print(f"Unmapped rows: {len(unmapped)}")
print(unmapped)
```

```python
moa_df.to_csv("drugs/moa_8_cui.tsv", sep="\t", index=False)
```
