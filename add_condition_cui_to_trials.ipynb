```python
import pandas as pd
```

```python
trial_df = pd.read_csv("drugs/trials_umls.tsv", sep = "\t")
trial_df.head()
```

```python
annotation_df = pd.read_csv("trial_condition_annotated - trials_cui.tsv", sep = "\t")
annotation_df.head()
```

```python
trial_df_extended = trial_df.merge(
    annotation_df, 
    left_on="Medical Condition", 
    right_on="condition", 
    how="left"
)
```

```python
trial_df_extended.to_csv("drugs/trials_umls_extended.tsv", sep="\t", index=False)
```
