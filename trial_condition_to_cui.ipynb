```python
import umls_api
import yaml
import pandas as pd
from tqdm import tqdm
```

```python
with open("config.yaml", "r") as stream:
    try:
        PARAM = yaml.safe_load(stream)
    except yaml.YAMLError as exc:
        print(exc)

umls_token = PARAM["umls_token"]
```

```python
trial_df = pd.read_csv("drugs/trials_umls.tsv", sep = "\t")
trial_df.head()
```

```python
condition_list = trial_df["Medical Condition"].unique().tolist()
```

```python
umls_api.search("Acne Vulgaris", umls_token, amount_of_results = 1)
```

```python
content = "condition\tcondition_cui\tcondition_preferred_umls_name\n"
for c in tqdm(condition_list):
  results = umls_api.search(c, umls_token, amount_of_results = 1)

  if len(results) == 1:
    cui = results[0]["cui"]
    name = results[0]["name"]
    content += f"{c}\t{cui}\t{name}\n"

with open('trials_cui.tsv', 'w') as file:
  file.write(content)
```
